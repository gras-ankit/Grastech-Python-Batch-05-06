```python
url = '''https://www.grastech.in/'''
import requests as r
response = r.get(url)
response
```

```python
html_data = response.content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html_data)
```

```python
type(soup)
```

```python
list(soup)[:2]
```

```python
all_reviews_message = []
all_names = []
all_designations = []
all_images = []
for i in soup.find_all('div',class_ = True):
    if i['class'][0] == 'career_card':
        for index,j in enumerate(i.find_all('div',class_ = True)):
            
            if j['class'][0] == 'flexibility-card__content':
                if index == 5:
                    # print(j.text.splitlines()[1].strip())
                    review = j.text.splitlines()[1].strip()
                    all_reviews_message.append(review)
                ...
    if i['class'][0] == 'col-lg-8':
        temp_data = i.text.splitlines()
        if len(temp_data) == 3:
            _,name,designation = temp_data
            # print(name)
            all_names.append(name)
            all_designations.append(designation)

    if i['class'][0] == 'col-lg-4':
        temp_data = i
        for k in i.find_all('img',src = True):
            # print(k['src'])
            all_images.append(k['src'])

# print(len(all_reviews_message))
# print(len(all_names))
# print(len(all_designations))

master_data = {'Name':all_names,'Job-Designation':all_designations,"Reviews-Message":all_reviews_message,'Image':all_images}

import pandas as pd
table = pd.DataFrame(master_data)
table
```

```python
table.to_excel('Grastech_reviews.xlsx')
print('Done')
```
