```python
def msg():
    name = 'India'

    def city_msg():
        nonlocal name
        name = 'Noida'
        print(f'Msg 2: I love {name}')

    city_msg()
    print(f'Msg 1: I Love {name}')

    
```

```python
t = 1
def call_me():
    global t
    print(f'Calling!! time: {t}')
    t += 1
    call_me()
```

```python
def facto(n):
    ans = 1
    for i in range(1,n+1):
        ans *= i
    return ans
```

```python
facto(40)
```

```python
def rec_facto(n):
    if n == 1:
        return n
    else:
        return n*rec_facto(n-1)
```

```python
rec_facto(5)
```

```python
facto(500)
```

```python
def rec_sum(n):
    if n == 1:
        return n
    else:
        return rec_sum(n-1)+n
```

```python
rec_sum(100)
```
