# Multilevel Inheritance

```python
class Grand_Father:
    def blind_money(self):
        print('')
        
    def land(self):
        print('Acers of Lands!!')

class Father(Grand_Father):
    def builders(self):
        print('Lot\'s of Plots, Properties, flats!!')
    def cars(self):
        print('Audi, BMW')

class Son(Father):
    def mind(self):
        print('I can use my mind!!')
    def hardwoking(self):
        print('Hardworking Person')

class Grand_Son(Son,Nokia):
    def nothing(self):
        print('I have Nothing!!')

```

```python
deepak = Grand_Son()
```

```python
deepak.nothing()
```

```python
deepak.blind_money()
```

```python
class God_Father:
    def super_magic(self):
        print('I can do anything!!')

class Grand_Father(God_Father):
    def blind_money(self):
        print('')
        
    def land(self):
        print('Acers of Lands!!')

class Father(God_Father):
    def builders(self):
        print('Lot\'s of Plots, Properties, flats!!')
    def cars(self):
        print('Audi, BMW')

class Son(God_Father):
    def mind(self):
        print('I can use my mind!!')
    def hardwoking(self):
        print('Hardworking Person')

class Grand_Son(God_Father):
    def nothing(self):
        print('I have Nothing!!')

```

```python
class Beta_1:
    def version(self):
        print('This is Version 1')

class Beta_2(Beta_1):
     def version(self):
        print('This is Version 2')

class Beta_3(Beta_2):
     def version(self):
        print('This is Version 3')
        super().version()
```

```python
class Animal:
    def speak(self):
        return 'I am an Animal, and can Speak!!'

class Cat(Animal):
    def speak(self):
        return('Meow Meow!!')

class Dog(Animal):
    def speak(self):
        return('Bark! Bark!!')
```
