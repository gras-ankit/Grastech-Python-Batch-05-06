```python
# and, or, not
```

```python
# and:- agar saage conditions True hain: to Final :- True
# And op ke dath agar koi bhi condition False hai to Final False
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a<b) and ('Vinayak' in students)
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a<b) and ('Vinayak' not in students) and (False == False)
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a<b) and (('Vinayak' in students) and (False != False))
```

```python
# or:- koi bhi True to Final True hoga
# Jab Saare False Honge to Final False
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a<b) or ('Vinayak' in students) or (False != False)
  # T                 F                     F
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a<b) or (('Vinayak' not in students) and (False != False))
  # T                 T                        F
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

(a>b) or (('Vinayak' not in students) and (False != False))
  # F                 T                        F
```

```python
# not:- reverse the result
```

```python
a = 345
b = 456
students = ['Ajay','Mayank','Karan','Ravi']

not((a>b) or (('Vinayak' not in students) and (False != False)))
  # F                 T                        F
```
